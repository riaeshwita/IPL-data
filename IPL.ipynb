```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#reading csv file from pandas
import pandas as pd
df = pd.read_csv(r"C:\Users\riapi\OneDrive\Documents\ipl.csv")
df
```

```python
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
                     'Mumbai Indians', 'Delhi Daredevils', 'Kings XI Punjab',
                     'Royal Challengers Bangalore' , 'Sunrisers Hyderabad']
df = df[(df['bat_team'].isin(consistent_teams)) & (df['bowl_team'].isin(consistent_teams))]
print("After removing in consistent teams: {}".format(df.shape))
df
```

```python
#removing the first 5 overs data in every match - current over must be greater than 5.0
print("Before removing in consistent teams: {}".format(df.shape))
df = df[df['overs']>=5.0]
print("After removing in consistent teams: {}".format(df.shape))
```

```python
df
```

```python
df.head()
```

```python
#converting date from string to Date-Time
from datetime import datetime

print("Before converting : {}".format(type(df.iloc[0,0])))
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
print("After converting : {}".format(type(df.iloc[0,0])))
```

```python
df.dtypes
```

```python
#Encoding:  converting string values to numerical values
#String values also called as CATEGORICAL VALUES(2 types)
#1: Nominal values: string values but no rank(rank cannot be assigned)
#    ->Onehotencoding - Scikitlearn library
#    ->dummy encoding - Pandas
#2: Ordinal values: has order/rank
#   ->Label encoding - Scikitlearn
#   ->Mapping

#dummy encoding: creates a separate column for unique value, then assign a boolean value
```

```python
#bat_team and bowl_team are nominal values
#to convert them we use Dummy Encoding

encoded_df = pd.get_dummies(data=df, columns=['bat_team','bowl_team'], dtype=int)
encoded_df
```

```python
encoded_df.info()
```

```python
encoded_df.columns
```

```python
encoded_df = encoded_df[['date','bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','overs','runs', 'wickets','runs_last_5', 'wickets_last_5',
       'total']]
encoded_df
```

```python
#Splitting the data into train and test set

x_train = encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year <= 2016]
x_test = encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year >= 2017]

y_train = encoded_df[encoded_df['date'].dt.year <= 2016]['total'].values
y_test = encoded_df[encoded_df['date'].dt.year >= 2017]['total'].values

#Removing the 'date' column
x_train.drop(labels='date', axis=True, inplace=True)
x_test.drop(labels='date', axis=True, inplace=True)

print("Training set: {} anf Testing set {} ".format(x_train.shape, x_test.shape))
print("Training set: {} anf Testing set {} ".format(y_train.shape, y_test.shape))
```

```python
#Regression
#1: Linear Regression: plot data using scatter plot, find best fitting linear line
#2: Decision Tree Regressor: if and else type
#3: Random Forest Regressor: multiple trees
```

```python
from sklearn.linear_model import LinearRegression

#model/object
linear_regressor = LinearRegression()

#training
linear_regressor.fit(x_train, y_train)
```

```python
#Predicting results - x_test
#actual value will be in y_test
y_pred_lr = linear_regressor.predict(x_test)
y_pred_lr  #x_test
```

```python
#model evaluation
#less mae -> better result

from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
print("Model Evaluation")
print("Mean Absolute Error(MAE) : {} ".format(mae(y_test,y_pred_lr)))
print("Mean Squared Error(MSE) : {} ".format(mse(y_test,y_pred_lr)))
print("Root Mean Square Erroe(RMSE) : {} ".format(np.sqrt((mse(y_test,y_pred_lr)))))
```

```python
#checking score of model
#higher score better results

print("Accuracy of LR on Training set : {:.2f} ".format(linear_regressor.score(x_train,y_train)))
print("Accuracy of LR on Testing set : {:.2f} ".format(linear_regressor.score(x_test,y_test)))
```

```python
#Decision Tree

from sklearn.tree import DecisionTreeRegressor

decision_regressor = DecisionTreeRegressor()

#training
decision_regressor.fit(x_train, y_train)
```

```python
#prediction
y_pred = decision_regressor.predict(x_test)
y_pred
```

```python
#model evaluation
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
print("Model Evaluation")
print("Mean Absolute Error(MAE) : {} ".format(mae(y_test,y_pred)))
print("Mean Squared Error(MSE) : {} ".format(mse(y_test,y_pred)))
print("Root Mean Square Erroe(RMSE) : {} ".format(np.sqrt((mse(y_test,y_pred)))))
```

```python
#checking score
print("Accuracy of LR on Training set : {:.2f} ".format(decision_regressor.score(x_train,y_train)))
print("Accuracy of LR on Testing set : {:.2f} ".format(decision_regressor.score(x_test,y_test)))
```

```python
#Random Forest

from sklearn.ensemble import RandomForestRegressor

random_regressor = RandomForestRegressor()
random_regressor.fit(x_train, y_train)
```

```python
#Taking values from user to make predictions

def predict_score(batting_team='Chennai Super Kings', bowling_team='Mumbai Indians', overs=5.1, runs=50, wickets=0,run_in_prev_5=0,wickets_in_prev_5=0):
    temp_array = list()
    
    #Batting team
    if batting_team == 'Chennai Super Kings':
        temp_array = temp_array + [1,0,0,0,0,0,0,0]
    elif batting_team == 'Delhi Daredevils':
        temp_array = temp_array + [0,1,0,0,0,0,0,0]
    elif batting_team == 'Kings XI Punjab':
        temp_array = temp_array + [0,0,1,0,0,0,0,0] 
    elif batting_team == 'Kolkata Knight Riders':
        temp_array = temp_array + [0,0,0,1,0,0,0,0]
    elif batting_team == 'Mumbai Indians':
        temp_array = temp_array + [0,0,0,0,1,0,0,0]
    elif batting_team == 'Rajasthan Royals':
        temp_array = temp_array + [0,0,0,0,0,1,0,0] 
    elif batting_team == 'Royal Challengers Bangalore':
        temp_array = temp_array + [0,0,0,0,0,0,1,0]  
    elif batting_team == 'Sunrisers Hyderabad':
        temp_array = temp_array + [0,0,0,0,0,0,0,1]      
        
   #Bowling team
    if bowling_team == 'Chennai Super Kings':
        temp_array = temp_array + [1,0,0,0,0,0,0,0]
    elif bowling_team == 'Delhi Daredevils':
        temp_array = temp_array + [0,1,0,0,0,0,0,0]
    elif bowling_team == 'Kings XI Punjab':
        temp_array = temp_array + [0,0,1,0,0,0,0,0] 
    elif bowling_team == 'Kolkata Knight Riders':
        temp_array = temp_array + [0,0,0,1,0,0,0,0]
    elif bowling_team == 'Mumbai Indians':
        temp_array = temp_array + [0,0,0,0,1,0,0,0]
    elif bowling_team == 'Rajasthan Royals':
        temp_array = temp_array + [0,0,0,0,0,1,0,0] 
    elif bowling_team == 'Royal Challengers Bangalore':
        temp_array = temp_array + [0,0,0,0,0,0,1,0]  
    elif bowling_team == 'Sunrisers Hyderabad':
        temp_array = temp_array + [0,0,0,0,0,0,0,1]  
        
    temp_array = temp_array + [overs,runs,wickets,run_in_prev_5,wickets_in_prev_5]
    
    temp_array = np.array([temp_array])
    print(temp_array)
    
    return int(linear_regressor.predict(temp_array))
```
